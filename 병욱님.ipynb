```python
# 년도별 5대 범죄 총 합계만 표시
y_sum = []
for x in range(7):
    yn = x + 4
    yy = str(2010 + int(yn))
    yy_1 = str(2010.1 + int(yn))
    y_sum.append(yy)
    y_sum.append(yy_1)

print(y_sum)

# str()을 사용하면 숫자를 문자열로 변경 가능
# str(y_sum)

y_sum.insert(0, "자치구별(2)")
df_2 = df_1[y_sum]
df_2 = df_2.drop(index = 1)
df_2
```

```python
# 0은 발생, 1은 검거
df_2 = df_2.reset_index(drop = True)
df_2.head()
```

```python
df_2 = df_2.drop(index = 0)
df_2.head()
```

```python
df_2.head()
```

```python
df_2.columns = ["자치구별", "2014년_발생수", "2014년_검거수", "2015년_발생수", "2015년_검거수", "2016년_발생수", "2016년_검거수", "2017년_발생수", "2017년_검거수", "2018년_발생수", "2018년_검거수", "2019년_발생수", "2019년_검거수", "2020년_발생수", "2020년_검거수"]
df_2
```

```python
df_3 = df_2.melt(id_vars = "자치구별", value_vars = df_2.columns[1:], value_name = "수치", var_name = "년도별 발생 및 검거수")
df_3
```

```python
df_3["연도"] = pd.to_numeric(df_3["년도별 발생 및 검거수"].str[:4], errors = 'coerce')
df_3
```

```python
df_3["년도별 발생 및 검거수"].str[6:]
```
